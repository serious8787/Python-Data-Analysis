# 项目：评估和清理英国电商公司销售数据

## 分析目标

此数据分析的目的是，根据市场销售数据，挖掘畅销产品，以便制定更有效的市场策略来提升营收。

本实战项目的目的在于练习评估数据干净和整洁度，并且基于评估结果，对数据进行清洗，从而得到可供下一步分析的数据。

## 简介

原始数据集记录了一家英国在线零售公司在2010年12月1日至2011年12月9日期间的所有交易情况，涵盖了该公司在全球不同国家和地区的业务数据。该公司主要销售覆盖各个场景的礼品，包括但不限于生日礼品、结婚纪念品、圣诞礼品等等。该公司的客户群体主要包括批发商和个人消费者，其中批发商占据了相当大的比例。

数据每列的含义如下：
- `InvoiceNo`: 发票号码。6位数，作为交易的唯一标识符。如果这个代码以字母“c”开头，表示这笔交易被取消。
- `StockCode`: 产品代码。5位数，作为产品的唯一标识符。
- `Description`: 产品名称。
- `Quantity`: 产品在交易中的数量。
- `InvoiceDate`: 发票日期和时间。交易发生的日期和时间。
- `UnitPrice`: 单价。价格单位为英镑（£）。
- `CustomerID`: 客户编号。5位数，作为客户的唯一标识符。
- `Country`: 国家名称。客户所居住的国家的名称。

## 清理数据

```python
#生成資料副本:方便區分原始和清理後數據，用 copy方法
cleaned_data = original_data.copy()
```

```python
#轉換類型為'日期時間類型':用函數 pd.to_datetime(提取某列)
cleaned_data['InvoiceDate'] = pd.to_datetime(cleaned_data['InvoiceDate'])  #把'InvoiceDate'列之字符類型轉變為datetime類型
cleaned_data['InvoiceDate']
```

```python
#一般轉換類型:用astype方法
cleaned_data['CustomerID'] = cleaned_data['CustomerID'].astype(str)  #把'CustomerID'列之字符類型轉變為object類型
cleaned_data['CustomerID']
```

```python
#刪除字符串頭尾 :用 str.slice(起始索引,結束索引) -- 頭留尾不留
cleaned_data['CustomerID'] = cleaned_data['CustomerID'].str.slice(0, 5)  #(0, 5) 結果同  (0, -2倒數第二個以前保留)
cleaned_data['CustomerID']  #刪除掉str中的'.0'
```

```python
#刪除空缺觀察值:用 dropna(subset=列名)
cleaned_data.dropna(subset='Description', inplace=True)
```

```python
  #確認使否已刪除(不再存有缺失值)
cleaned_data['Description'].isnull().sum()  #加總為0，確實已刪除
```

```python
#替換值(解決不一致):用 replace({'舊名':'替代名', ...})
cleaned_data['Country'] = cleaned_data['Country'].replace({'USA': 'United States'})
```

```python
    #確認使否已刪除(不再存有缺失值):len(df)可以計算出df行數
len(cleaned_data[cleaned_data['Country'] == 'USA'])  #提取值含'USA'之df 行數為0
```

```python
cleaned_data['Country'] = cleaned_data['Country'].replace({'UK': 'United Kingdom', 'U.K.': 'United Kingdom'})
```

```python
len(cleaned_data[cleaned_data['Country'] == 'UK'])  #提取值含'UK'之df 行數為0
len(cleaned_data[cleaned_data['Country'] == 'U.K.'])  #提取值含'U.K.'之df 行數為0
```

```python
#保留達條件之觀察值
cleaned_data = cleaned_data[cleaned_data['Quantity'] >= 0]  #保留'Quantity'值>= 0之行
```

```python
  #確保不符條件皆已刪除
len(cleaned_data[cleaned_data['Quantity'] < 0])
```

```python
cleaned_data = cleaned_data[cleaned_data['UnitPrice'] >= 0]  #保留'UnitPrice'值>= 0之行
```

```python
  #確保不符條件皆已刪除
len(cleaned_data[cleaned_data['UnitPrice'] < 0])
```

## 保存清理后的数据

```python
#保存並輸出數據:用 to_csv('取文件名') ，會儲存至同一文件夾下
cleaned_data.to_csv('e_commerce_cleaned.csv', index=False)  #儲存時刪除無意義索引，否則讀取時會被當成列(需再set_index)
```
